```r
library(pracma)
```

```r
f <- c(0, 0, .1, .2, .25, .2, .25, .3, .35, .43, .45, .5, .55, .5, .4, .425, .45, .425, .4, .35, .3, .25, .225, .2, .1, 0, 0)
g <- c(0, 0, .025, .1, .2, .175, .2, .25, .25, .3, .32, .35, .375, .325, .3, .275, .275, .25, .225, .225, .2, .175, .15, .15, .05, 0, 0)

ones <- rep(1, length(g))

g <- g - 0.025 * ones

x <- 0.1 * seq(1, length(f))
xf <- seq(0.1, tail(x, n=1), 0.01)

ff = interp1(x, f, method='cubic')
gg = interp1(x, g, method='cubic')

plot(xf[1:length(ff)],ff,col='black',lwd=2, type="l")
points(xf[1:length(ff)],ff, col="blue", pch=19)
lines(xf[1:length(gg)],gg,col='black',lwd=2, type="l")
points(xf[1:length(gg)],gg, col="red", pch=19)
```

```r
dx <- 0.01
L <- 1000
x <- seq(0,L,dx)
n <- length(x)
nquart <- as.integer(floor(n/4))

f <- rep(0, length(x))

f[(nquart+1):(3*nquart)] <- 1

A0 <- sum(f * rep(1, length(x))) * dx * 2 / L
fFS <- A0/2 * rep(1, length(f))

j <- 0
for (k in 1:100) {
    Ak <- sum(f * cos(2*pi*k*x/L)) * dx * 2 / L
    Bk <- sum(f * sin(2*pi*k*x/L)) * dx * 2 / L
    fFS <- fFS + Ak*cos(2*k*pi*x/L) + Bk*sin(2*k*pi*x/L)
}
    
plot(x, f, type="l", lwd=2)
lines(x, fFS, col="red", lwd=2)
```

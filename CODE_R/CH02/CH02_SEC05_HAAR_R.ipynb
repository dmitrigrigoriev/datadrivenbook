```r
x <- seq(0,1,0.001)
n <- length(x)

n2 = as.integer(floor(n/2))
n4 = as.integer(floor(n/4))

f10 <- rep(0, length(x))
f10[1:n2] <- 1
f10[n2:length(f10)] <- -1

f21 <- rep(0, length(x))
f21[1:n4] <- 1
f21[n4:n2] <- -1
f21 <- f21 * sqrt(2)

f22 <- rep(0, length(x))
f22[n2:(n2+n4)] <- 1
f22[(n2+n4):length(f22)] <- -1
f22 = f22 * sqrt(2)

x <- append(x, c(-1,0), after=0)
x <- append(x, c(1, 2), after=length(x))

f10 <- append(f10, c(0, 0), after=0)
f10 <- append(f10, c(0, 0), after=length(f10))

f21 <- append(f21, c(0, 0), after=0)
f21 <- append(f21, c(0, 0), after=length(f21))

f22 <- append(f22, c(0, 0), after=0)
f22 <- append(f22, c(0, 0), after=length(f22))
```

```r
par(mfrow=c(3,1))

plot(x, f10, type="l", lwd="2")
plot(x, f21, type="l", lwd="2")
plot(x, f22, type="l", lwd="2")
```

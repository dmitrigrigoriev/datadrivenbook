```r
library(mrbsizeR)
library(sparr)
library(EBImage)
```

```r
#Load Image
#A = load.image("../DATA/jelly.jpg")
A = readImage("../DATA/jelly.jpg")


#Convert to grayscale image
#X = grayscale(A)
X = channel(A, "grey")

#Show Image
display(X)
```

```r
img_ff <- fft2d(X)

magntd <- sqrt(Re(img_ff)^2+Im(img_ff)^2)
img_fftsh <- fftshift(magntd)
display(log(img_fftsh),title="FFT")
```

```r
library(pracma)
```

```r
n <- 100
L <- 4
x <- linspace(0,L,n)
f <- x^2 # Parabola with 100 data points

M <- 20 # Polynomial degree

phi <- matrix(0, n, M)
for (j in 1:M) {
    phi[,j] = x^(j-1) # Build matrix A
}

plot(x,f,col='black', type="l", lwd="3")

par(mfrow=c(2,2))

for (j in 1:4){
  r <- rnorm(length(x), 0, 1)
  fn <- x^2 + 0.1 * r
  an <- pinv(phi) %*% fn
  
  barplot(t(an))
}
```

```r
library(pracma)
```

```r
x <- linspace(0.2,5,50)
y <- 1.0 / x

x2 <- linspace(0.2,5,50)
y2 <- 1.0 / x2 + 0.5 * rnorm(length(x2), 0, 1)


y3 <- as.vector(replicate(5, y2)+2*rand(length(y2), 5)+1)
x3 <- as.vector(replicate(5, x))
```

```r
plot(x, y, type="l", col="black")
points(x2, y2, pch=16, col="green")
points(x3, y3, pch=16, col="magenta")
rect(0.5, 0.4, 1.4, 2, col="grey", density=70)
```

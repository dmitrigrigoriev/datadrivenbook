```r
library(MASS)
```

```r
H <- read.csv("../DATA/housing.data", header = FALSE, sep = "", dec = ".")

b <- as.matrix(H[,ncol(H)])
A <- as.matrix(H[,-ncol(H)])
```

```r
# Solve Ax=b using SVD
# Note that the book uses the Matlab-specific "regress" command

res <- svd(A)

S <- diag(res$d)
U <- res$u
V <- res$v

x = V %*% ginv(S) %*% t(U) %*% b

d <- 1:length(b)

par(mfrow = c(1, 1), mar=c(3, 3, 3, 0))

solution <- A%*%x

plot(d, b, col="black", lty = 1, type = "b", pch=19, xlab="Neighborhood", ylab="Median Home Value [$1k]")
lines(d, solution,col="red")
points(d, solution, pch = 19, col = "red")  ## only points
grid()
legend("topleft", legend=c("Housing Value", "Regression"), 
       col=c("black", "red"), lwd=1, lty=c(1,1), 
       pch=c(NA,19))


sort_indx <- order(b)
coords <- cbind(rbind(sort_indx))

plot(d, b[coords], col="black", lty = 1, type = "b", pch=19, xlab="Neighborhood", ylab="Median Home Value [$1k]")
lines(d, solution[coords],col="red")
points(d,solution[coords], pch = 19, col = "red")  ## only points
grid()
legend("topleft", legend=c("Housing Value", "Regression"), 
       col=c("black", "red"), lwd=1, lty=c(1,1), 
       pch=c(NA,19))
```
